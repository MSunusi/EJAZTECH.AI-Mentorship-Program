```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import recall_score, precision_score, accuracy_score
from sklearn.impute import SimpleImputer
import seaborn as sns
```

```python
# load the dataset
data = pd.read_csv("/content/sample_data/Assignment-2_Data.csv")
```

```python
# Set Input and output features
X = data.drop(columns = ['y', 'Id'])
Y = data['y']
```

```python
#Display input X
X
```

```python
# Display output
Y
```

```python
# handle missing value
imputer = SimpleImputer(strategy='most_frequent')
imputed_X = imputer.fit_transform(X)
d_X = pd.DataFrame(imputed_X, columns=X.columns)
```

```python
d_X
```

```python
# Apply onehot encoding for non numeric features
d_X_encoded = pd.get_dummies(d_X)
```

```python
#Display
d_X_encoded
```

```python
# Convert True/False to 1/0
d_X_encoded = d_X_encoded.astype(int)

# Scale
scaler = StandardScaler()
d_X_scaled = scaler.fit_transform(d_X_encoded)
```

```python
#Split train and test set
X_train,X_test,Y_train,Y_test=train_test_split(d_X_scaled,Y, test_size=0.2)
```

```python
# train the model
LR=LogisticRegression(max_iter= 20000)
LR.fit(X_train,Y_train)
```

```python
# Make prediction
predicted=LR.predict(X_test)
predicted
```

```python
# Evaluate the model Performance
acc_score = accuracy_score(Y_test, predicted)
print(f"Accuracy score: {acc_score}")
```
